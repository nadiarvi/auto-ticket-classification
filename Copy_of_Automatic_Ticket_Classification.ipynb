```python
import bllipparser
from bllipparser import RerankingParser
# un-comment the code when running for the first time
rrp = RerankingParser.fetch_and_load('WSJ-PTB3', verbose=True)
# rrp.simple_parse("It's that easy.")
```

```python
sents = sent_tokenize(df['complaint'][17])
sents
```

```python
for sent in sents:
  parse = rrp.simple_parse(sent)
  tree = bllipparser.Tree(parse)
  subtree_labels = [subtree.label for subtree in tree[0].subtrees()]

  parsed = []
  if 'S' in subtree_labels:
    parsed += [' '.join(subtree.tokens()) for subtree in tree[0].subtrees() if subtree.label == 'S']
    text = '. '.join(parsed)
  else:
    text = sent

  print(text)
```

```python
def parse_complaint(text):
  sents = sent_tokenize(text)

  res = []
  for sent in sents:
    parse = rrp.simple_parse(sent)
    tree = bllipparser.Tree(parse)
    subtree_labels = [subtree.label for subtree in tree[0].subtrees()]

    parsed = []
    if 'S' in subtree_labels:
      parsed += [' '.join(subtree.tokens()) for subtree in tree[0].subtrees() if subtree.label == 'S']
      res.append('. '.join(parsed))
    else:
      res.append(sent)

    # print(res)
  parsed_text = '. '.join(res)
  return parsed_text
```

```python
for i in range(5):
  pprint.pprint(parse_complaint(df['complaint'][i]), compact = True)
```

```python
df = pd.read_csv(root + 'clean_data.csv')
df.head()
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords = list(STOP_WORDS)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text = df['complaint'][0]
text
```

```python
doc = nlp(text)
tokens = [token.text for token in doc]
print(tokens)
```

```python
punctuation = punctuation + '\n'
punctuation
```

```python
word_frequencies = {}
for word in doc:
  if word.text.lower() not in stopwords:
    if word.text.lower() not in punctuation:
      if word.text not in word_frequencies.keys():
        word_frequencies[word.text] = 1
      else:
        word_frequencies[word.text] += 1
```

```python
max_frequency = max(word_frequencies.values())
```

```python
for word in word_frequencies.keys():
  word_frequencies[word] = word_frequencies[word]/max_frequency
```

```python
sentence_tokens = [sent for sent in doc.sents]
```

```python
sentence_scores = {}
for sent in sentence_tokens:
  for word in sent:
    if word.text.lower() in word_frequencies.keys():
      if sent not in sentence_scores.keys():
        sentence_scores[sent] = word_frequencies[word.text.lower()]
      else:
        sentence_scores[sent] += word_frequencies[word.text.lower()]
```

```python
from heapq import nlargest
```

```python
select_length = int(len(sentence_tokens)*0.7)
select_length
```

```python
summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
summary
```
